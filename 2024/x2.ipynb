```python
import os

def replace_labels_in_files(directory):
    """Replace labels 80 with 0 and 81 with 1 in all text files within a directory."""
    for filename in os.listdir(directory):
        if filename.endswith(".txt"):
            filepath = os.path.join(directory, filename)
            
            with open(filepath, 'r') as file:
                lines = file.readlines()
            
            with open(filepath, 'w') as file:
                for line in lines:
                    # Replace the labels
                    new_line = line.replace("80", "0").replace("81", "1")
                    file.write(new_line)
            print(f"Processed file: {filename}")
```
